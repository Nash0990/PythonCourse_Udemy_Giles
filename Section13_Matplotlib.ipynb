### Creating a Random walk
Imagine that you are standing on a grid and you have a coin. Every time you toss the coin, you get heads then you go up and move forward, incase you get tails you go down and move forward. The path traversed by this action is said to be compleatly random and dependent on the outcome of the coin toss.

```python
step_option=[1,-1]
step_choice=choice(step_option)
print(step_choice)
```

```python
walk=[]
walk.append(step_choice)
for step in range(1,1000):
    next_step=walk[step-1]+choice(step_option)
    walk.append(next_step)
print(walk)
```

```python
plt.plot(walk)
plt.show()
```

```python
def rand_walk(step_num):
    walk=[]
    step_choice=choice([1,-1])
    walk.append(step_choice)
    for i in range(1,step_num):
        step_choice2=choice([1,-1])
        next_step=walk[i-1]+step_choice2
        walk.append(next_step)
    return walk
```

```python
random_w=rand_walk(1000)
print(len(random_w))
```

```python
def plot_walk(walk):
    x=list(range(1,len(walk)+1))
    
    plt.plot(x,walk)
    plt.xlabel('Number of steps')
    plt.ylabel('Distance from origin')
    plt.title('Our Random Walk')
    plt.show()
```

```python
plot_walk(rand_walk(1000))
```

### Function that creates, multiple random walks.

```python
def plot_walks(num,walk):
    lab_list=list(range(1,num+1))
    for i in range(0,num):
        x=list(range(1,(walk)+1))
        plt.plot(x,rand_walk(walk),label='Plot number' + ' ' + str(lab_list[i]))
        plt.xlabel('Number of steps')
        plt.ylabel('Distance from origin')
        plt.title('Our Random Walk')
        plt.legend(loc='lower left')
    plt.show()
```

```python
plot_walks(5,100000)
```

### Sierpinski Triangle
First tranformation
\begin{equation}
x_{n+1} =0.5x_{n}
\end{equation}
\begin{equation}
y_{n+1} =0.5y_{n}
\end{equation}

Second transformation

\begin{equation}
x_{n+1} =0.5x_{n} + 0.5
\end{equation}
\begin{equation}
y_{n+1} =0.5y_{n} + 0.5
\end{equation}


Third transformation

\begin{equation}
x_{n+1} =0.5x_{n} + 1
\end{equation}
\begin{equation}
y_{n+1} =0.5y_{n}
\end{equation}


```python
def trans_1(p):
    x = p[0]
    y = p[1]
    x1=x*0.5
    y1=y*0.5
    return x1,y1

def trans_2(p):
    x = p[0]
    y = p[1]
    x1=x*0.5 + 0.5
    y1=y*0.5 + 0.5
    return x1,y1

def trans_3(p):
    x = p[0]
    y = p[1]
    x1=x*0.5 + 1
    y1=y*0.5
    return x1,y1
```

```python
tansformations=[trans_1,trans_2,trans_3]
a1 = [0]
b1 = [0]
a,b = 0,0

for i in range(100000):
    trans=choice(tansformations)
    a,b=trans((a,b))
    a1.append(a)
    b1.append(b)
    
```

```python
plt.rc('figure',figsize=(16,16))
plt.plot(a1,b1,'o')
#plt.savefig('Section13_Matplotlib_fig.png')
```
